```python
from collections import deque

n = int(input())

array = [[] for _ in range(n + 1)]

while True:
    a, b = map(int, input().split())
    if a == -1 and b == -1:
        break
        
    array[a].append(b)
    array[b].append(a)
    
answer = []
answer_point = int(1e9)

for i in range(1, n + 1):
    
    now_point = -1
    
    visited = [False] * (n + 1)
    
    q = deque()
    q.append((i, 0))
    
    visited[i] = True
    
    while q:
        now, now_count = q.popleft()
        now_point = max(now_point, now_count)
        
        for k in array[now]:
            if visited[k] == False:
                visited[k] = True
                q.append((k, now_count + 1))
    
    answer_point = min(answer_point, now_point)
    answer.append((i, now_point))
    
count = 0
temp = []

for i in answer:
    nn, pp = i
    if pp == answer_point:
        count += 1
        temp.append(nn)

print(answer_point, count)
for i in temp:
    print(i, end = ' ')
```
