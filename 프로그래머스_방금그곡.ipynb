```python
from collections import defaultdict

def solution(m, musicinfos):
    answer = ''
    d = defaultdict(int)
    m = m.replace("C#", "c")
    m = m.replace("D#", "d")
    m = m.replace("F#", "f")
    m = m.replace("G#", "g")
    m = m.replace("A#", "a")
    
    for i in musicinfos:
        s = ''
        
        start, end, title, mm = i.split(',')
        mm = mm.replace("C#", "c")
        mm = mm.replace("D#", "d")
        mm = mm.replace("F#", "f")
        mm = mm.replace("G#", "g")
        mm = mm.replace("A#", "a")
        
        start_h, start_m = start.split(":")
        end_h, end_m = end.split(":")
        
        start_h = int(start_h)
        start_m = int(start_m)
        end_h = int(end_h)
        end_m = int(end_m)
        
        temp = 0
        
        if end_h > start_h:
            end_h -= 1
            end_m += 60
            temp = (end_h - start_h) * 60 + end_m - start_m
            
        else:
            temp = (end_h - start_h) * 60 + end_m - start_m
        
        if temp <= len(mm):
            s += mm[:temp]
        
        else:
            while temp > len(mm):
                s += mm
                temp -= len(mm)
            s += mm[:temp]
        
        d[s] = title
        
    for j in d.keys():
        if m in j:
            if answer == '':
                answer = d[j]
            
            else:
                check = 0
                for k in d.items():
                    if k[1] == answer:
                        check = len(k[0])
                if check < len(j):
                    answer = d[j]

    if answer == '':
        return "(None)"
    
    return answer
```
