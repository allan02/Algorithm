## 17136번, 색종이 붙이기

```python
def function(x, y, cnt):
    global ans
    if y >= 10:
        ans = min(ans, cnt)
        return
    
    if x >= 10:
        function(0, y + 1, cnt)
        return
    
    if array[x][y] == 1:
        for k in range(5):
            if paper[k] == 5:
                continue
                
            if x + k >= 10 or y + k >= 10:
                continue
                
            flag = 0
            for i in range(x, x + k + 1):
                for j in range(y, y + k + 1):
                    if array[i][j] == 0:
                        flag = 1
                        break
                if flag:
                    break
                        
            if not flag:
                for i in range(x, x + k + 1):
                    for j in range(y, y + k + 1):
                        array[i][j] = 0
                        
                paper[k] += 1
                function(x + k + 1, y, cnt + 1)
                paper[k] -= 1
                
                for i in range(x, x + k + 1):
                    for j in range(y, y + k + 1):
                        array[i][j] = 1
                            
    else:
        function(x + 1, y, cnt)
                
                
array = [list(map(int, input().split())) for _ in range(10)]
paper = [0 for _ in range(5)]

ans = int(1e9)

function(0, 0, 0)

if ans == int(1e9):
    print(-1)
else:
    print(ans)

                
        
    
```
