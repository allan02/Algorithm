```python
dx = [0, 1, 0, -1]
dy = [-1, 0, 1, 0]

n = int(input())

array = []

for _ in range(n):
    array.append(list(map(int, input().split())))

direction = 0
count = 1
count_check = 0

start_x = n // 2
start_y = n // 2

answer = 0
check = False

while True:
    for _ in range(count):
        
        if start_x == 0 and start_y == 0:
            check = True
            break
            
        pre_x = start_x
        pre_y = start_y
            
        start_x += dx[direction]
        start_y += dy[direction]
        
        now = array[start_x][start_y]
        
        temp_d1 = direction - 1
        
        if temp_d1 < 0:
            temp_d1 = 3
        
        temp_d2 = direction + 1
        
        if temp_d2 > 3:
            temp_d2 -= 4
            
        # 5%
        temp_x = start_x + 2 * dx[direction]
        temp_y = start_y + 2 * dy[direction]
        
        if 0 <= temp_x < n and 0 <= temp_y < n:
            array[temp_x][temp_y] += int(array[start_x][start_y] * 0.05)
            now -= int(array[start_x][start_y] * 0.05)
        else:
            answer += int(array[start_x][start_y] * 0.05)
            now -= int(array[start_x][start_y] * 0.05)
            
        # 10%
        temp_x = start_x + dx[direction] + dx[temp_d1]
        temp_y = start_y + dy[direction] + dy[temp_d1]
        
        if 0 <= temp_x < n and 0 <= temp_y < n:
            array[temp_x][temp_y] += int(array[start_x][start_y] * 0.1)
            now -= int(array[start_x][start_y] * 0.1)
        else:
            answer += int(array[start_x][start_y] * 0.1)
            now -= int(array[start_x][start_y] * 0.1)
            
        temp_x = start_x + dx[direction] + dx[temp_d2]
        temp_y = start_y + dy[direction] + dy[temp_d2]
        
        if 0 <= temp_x < n and 0 <= temp_y < n:
            array[temp_x][temp_y] += int(array[start_x][start_y] * 0.1)
            now -= int(array[start_x][start_y] * 0.1)
        else:
            answer += int(array[start_x][start_y] * 0.1)
            now -= int(array[start_x][start_y] * 0.1)
            
        # 7%
        temp_x = start_x + dx[temp_d1]
        temp_y = start_y + dy[temp_d1]
            
        if 0 <= temp_x < n and 0 <= temp_y < n:
            array[temp_x][temp_y] += int(array[start_x][start_y] * 0.07)
            now -= int(array[start_x][start_y] * 0.07)
        else:
            answer += int(array[start_x][start_y] * 0.07)
            now -= int(array[start_x][start_y] * 0.07)
            
        temp_x = start_x + dx[temp_d2]
        temp_y = start_y + dy[temp_d2]
            
        if 0 <= temp_x < n and 0 <= temp_y < n:
            array[temp_x][temp_y] += int(array[start_x][start_y] * 0.07)
            now -= int(array[start_x][start_y] * 0.07)
        else:
            answer += int(array[start_x][start_y] * 0.07)
            now -= int(array[start_x][start_y] * 0.07)
            
        # 2%    
        
        temp_x = start_x + 2 * dx[temp_d1]
        temp_y = start_y + 2 * dy[temp_d1]
        
        if 0 <= temp_x < n and 0 <= temp_y < n:
            array[temp_x][temp_y] += int(array[start_x][start_y] * 0.02)
            now -= int(array[start_x][start_y] * 0.02)
        else:
            answer += int(array[start_x][start_y] * 0.02)
            now -= int(array[start_x][start_y] * 0.02)
            
        temp_x = start_x + 2 * dx[temp_d2]
        temp_y = start_y + 2 * dy[temp_d2]
        
        if 0 <= temp_x < n and 0 <= temp_y < n:
            array[temp_x][temp_y] += int(array[start_x][start_y] * 0.02)
            now -= int(array[start_x][start_y] * 0.02)
        else:
            answer += int(array[start_x][start_y] * 0.02)
            now -= int(array[start_x][start_y] * 0.02)
        
        # 1%
        temp_x = pre_x + dx[temp_d1]
        temp_y = pre_y + dy[temp_d1]
        
        if 0 <= temp_x < n and 0 <= temp_y < n:
            array[temp_x][temp_y] += int(array[start_x][start_y] * 0.01)
            now -= int(array[start_x][start_y] * 0.01)
        else:
            answer += int(array[start_x][start_y] * 0.01)
            now -= int(array[start_x][start_y] * 0.01)
            
        temp_x = pre_x + dx[temp_d2]
        temp_y = pre_y + dy[temp_d2]
        
        if 0 <= temp_x < n and 0 <= temp_y < n:
            array[temp_x][temp_y] += int(array[start_x][start_y] * 0.01)
            now -= int(array[start_x][start_y] * 0.01)
        else:
            answer += int(array[start_x][start_y] * 0.01)
            now -= int(array[start_x][start_y] * 0.01)
        
        # 알파 위치
        temp_x = start_x + dx[direction]
        temp_y = start_y + dy[direction]
        
        if 0 <= temp_x < n and 0 <= temp_y < n:
            array[temp_x][temp_y] += now
        else:
            answer += now
            
        array[start_x][start_y] = 0
        #print(array)
        
    
    if check:
        break
        
    count_check += 1 
    direction = (direction + 1) % 4
    
    if count_check % 2 == 0:
        count += 1
        count_check = 0
    
    
print(answer)

    
```
