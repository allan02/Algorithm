```python
from collections import deque

dx = [0, 0, 1, -1]
dy = [1, -1, 0, 0]

n, m = map(int, input().split())

answer = 0

array = []

for _ in range(n):
    array.append(list(map(int, input().split())))

def rotate(arr, n):
    temp = [[0] * n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            temp[i][j] = array[j][n - i - 1]
    return temp
    
while True:
    block = []
    visited = [[0] * n for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            if array[i][j] == -2 or array[i][j] == 0 or array[i][j] == -1:
                continue
                
            if visited[i][j] == 1:
                continue
                
            visited[i][j] = 1
                
            check = [[0] * n for _ in range(n)]
            now = array[i][j]
            
            q = deque()
            q.append((i, j))
            check[i][j] = 1
            
            count = 1
            count_0 = 0
            
            while q:
                x, y = q.popleft()
                
                for k in range(4):
                    nx = x + dx[k]
                    ny = y + dy[k]
                    
                    if 0 <= nx < n and 0 <= ny < n:
                        if check[nx][ny] == 0:
                            if array[nx][ny] == now:
                                visited[nx][ny] = 1
                                count += 1
                                check[nx][ny] = 1
                                q.append((nx, ny))
                                
                            elif array[nx][ny] == 0:
                                count += 1
                                count_0 += 1
                                check[nx][ny] = 1
                                q.append((nx, ny))
                                
            block.append((count, count_0, i, j))
                        
    if len(block) == 0:
        break
        
    block = sorted(block, key = lambda x : (-x[0], -x[1], -x[2], -x[3]))
    
    if block[0][0] < 2:
        break
    
    answer += (block[0][0] * block[0][0])
    
    start_x = block[0][2]
    start_y = block[0][3]
    
    v = array[start_x][start_y]
    
    q = deque()
    
    array[start_x][start_y] = -2
    q.append((start_x, start_y))
    
    while q:
        now_x, now_y = q.popleft()
        for k in range(4):
            nx = now_x + dx[k]
            ny = now_y + dy[k]
            
            if 0 <= nx < n and 0 <= ny < n:
                if array[nx][ny] == v or array[nx][ny] == 0:
                    array[nx][ny] = -2
                    q.append((nx, ny))
                    
                    
    for i in range(n):
        for j in range(n - 1, 0, -1):
            if array[j][i] == -1:
                continue
                
            if array[j][i] == -2:
                for k in range(j, -1, -1):
                    if array[k][i] == -1:
                        break
                    
                    if array[k][i] != -1 and array[k][i] != -2:
                        array[j][i] = array[k][i]
                        array[k][i] = -2
                        break
                
                
    array = rotate(array, n)
    
    
    for i in range(n):
        for j in range(n - 1, 0, -1):
            if array[j][i] == -1:
                continue
                
            if array[j][i] == -2:
                for k in range(j, -1, -1):
                    if array[k][i] == -1:
                        break
                    
                    if array[k][i] != -1 and array[k][i] != -2:
                        array[j][i] = array[k][i]
                        array[k][i] = -2
                        break
                
print(answer)

    
    

    
            
```
