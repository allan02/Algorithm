```python
def solution(dirs):
    answer = 0
    array = []
    dx = [1, -1, 0, 0]
    dy = [0, 0, -1, 1]
    now_x = 5
    now_y = 5
    
    for i in dirs:
        if i == 'R':
            temp_x = now_x
            temp_y = now_y
            
            now_x += dx[0]
            now_y += dy[0]
            
            if now_x > 10 or now_y > 10 or now_x < 0 or now_y < 0:
                now_x = temp_x
                now_y = temp_y
                continue
            
            if (temp_x, temp_y, now_x, now_y) not in array:
                array.append((temp_x, temp_y, now_x, now_y))
                array.append((now_x, now_y, temp_x, temp_y))
                
        if i == 'L':
            
            temp_x = now_x
            temp_y = now_y
            now_x += dx[1]
            now_y += dy[1]
            
            if now_x > 10 or now_y > 10 or now_x < 0 or now_y < 0:
                now_x = temp_x
                now_y = temp_y
                continue
            
            if (temp_x, temp_y, now_x, now_y) not in array:
                array.append((temp_x, temp_y, now_x, now_y))
                array.append((now_x, now_y, temp_x, temp_y))
                
        if i == 'D':
            
            temp_x = now_x
            temp_y = now_y
            now_x += dx[2]
            now_y += dy[2]
            
            if now_x > 10 or now_y > 10 or now_x < 0 or now_y < 0:
                now_x = temp_x
                now_y = temp_y
                continue
            if (temp_x, temp_y, now_x, now_y) not in array:
                array.append((temp_x, temp_y, now_x, now_y))
                array.append((now_x, now_y, temp_x, temp_y))
                
        if i == 'U':
            
            temp_x = now_x
            temp_y = now_y
            
            now_x += dx[3]
            now_y += dy[3]
            
    
            if now_x > 10 or now_y > 10 or now_x < 0 or now_y < 0:
                now_x = temp_x
                now_y = temp_y
                continue
            if (temp_x, temp_y, now_x, now_y) not in array:
                array.append((temp_x, temp_y, now_x, now_y))
                array.append((now_x, now_y, temp_x, temp_y))

    return len(array) // 2 
```
